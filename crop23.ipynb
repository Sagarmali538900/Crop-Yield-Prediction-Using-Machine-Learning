```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
crop_datagen = ImageDataGenerator(zoom_range=0.4, shear_range=.3,rotation_range=90, 
                              horizontal_flip=True, rescale=1/255)
```

```python
crop_data =crop_datagen.flow_from_directory(
    directory='crop_images/',
    target_size=(224,224),
    batch_size=3,
    class_mode="categorical")
```

```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()
model.add(Conv2D(16,(3,3),activation='relu',input_shape=(224,224,3)))
model.add(MaxPool2D())
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPool2D()) 
model.add(Conv2D(64,(5,5),activation='relu'))
model.add(MaxPool2D()) 
model.add(Flatten())
model.add(Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'], optimizer='Adam')
```

```python
model.fit(crop_data,epochs=40)
```

```python
def get_class(path):
    A=cv2.imread(path)
    A=cv2.resize(A,(224,224))
    A=A.reshape(1,224,224,3)/255
    print(model.predict(A))
    if model.predict(A).argmax()==0:
        return('jute')
    elif model.predict(A).argmax()==1:
        return('maize')
    elif model.predict(A).argmax()==2:
        return('rice')
    elif model.predict(A).argmax()==3:
        return('sugarcane')
    elif model.predict(A).argmax()==4:
        return('wheat')
    
    
    
        
```
